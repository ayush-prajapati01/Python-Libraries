```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q1
**Python program to draw bar plot of sex against survived for a dataset given in the url**<br>
<br>
Dataset: https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
def main():
    # Loading the Titanic dataset from the github URL
    url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
    df = pd.read_csv(url)
    
    print(f"The dataset is:\n{df.head()}\n")
    # Create a bar plot
    sns.barplot(data=df, x='sex', y='survived', hue='sex', saturation=0.8, palette="flare", legend=True)

    # Set plot labels and title
    plt.title('Survival by Sex')
    plt.xlabel('Sex')
    plt.ylabel('Survived')
    plt.show()
    

if __name__ == "__main__":
    main()
```
