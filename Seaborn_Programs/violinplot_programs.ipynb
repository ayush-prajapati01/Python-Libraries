```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q1
**Python program to draw a violin plot of sex against total_bill for a given dataset**
<br><br>
Dataset: `https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv`

```python
def main():
    # Loading the dataset using the link
    df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
    print(f"The columns of the dataset are: {df.columns}")
    print(f"\nThe violin plot is\n\n")
    plt.figure(figsize=(8,6))
    
    # Create a violin plot
    sns.violinplot(data=df, x='sex', y='total_bill', hue='sex', palette="flare", legend=True)
    
    # Set plot labels and title
    plt.title('Sex vs Total Bill')
    plt.xlabel('Sex')
    plt.ylabel('Total Bill')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(loc = "upper left")
    plt.show()
    

if __name__ == "__main__":
    main()
```

## Q2
**Python program to draw a violin plot of “species” against “sepal length” for a dataset**
<br><br>
Dataset: `https://github.com/mwaskom/seaborn-data/blob/master/iris.csv`


```python
def main():
    # Loading the dataset using the link
    df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
    print(f"The columns of the dataset are: \n{df}\n\n")
    print(f"The columns of the dataset are: {df.columns}")
    print(f"\nThe violin plot is\n\n")
    plt.figure(figsize=(9,6))
    
    # Create a violin plot
    sns.violinplot(data=df, x='species', y='sepal_length', hue='species', palette="deep", legend=True)
    
    # Set plot labels and title
    plt.title('Species vs Sepal Length')
    plt.xlabel('Species')
    plt.ylabel('Sepal Length')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.legend(loc = "upper left")
    plt.show()
    

if __name__ == "__main__":
    main()
```
