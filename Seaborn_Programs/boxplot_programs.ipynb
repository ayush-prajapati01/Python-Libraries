```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q1
**Write a program to draw box plot of life expectancy by continent for a data set given in a url** 
<br><br>
Dataset: `https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv`



```python
def main():
    # Loading the dataset using the link
    df = pd.read_csv('https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv')
    print(f"The columns of the dataset are: \n{df}\n\n")
    print(f"The columns of the dataset are: {df.columns}")
    print(f"\nThe box plot is\n\n")
    plt.figure(figsize=(13,7))
    
    # Create a box plot
    sns.boxplot(data=df, x='continent', y='lifeExp', hue='continent', palette="flare", legend=True)
    
    # Set plot labels and title
    plt.title('Continent vs Life Expectancy')
    plt.xlabel('Continent', labelpad=10)
    plt.ylabel('Life Expectancy', labelpad=12)
    plt.grid(axis='y', linestyle='-.', alpha=0.7)
    plt.legend(loc = "lower right")
    plt.show()
    

if __name__ == "__main__":
    main()
```
