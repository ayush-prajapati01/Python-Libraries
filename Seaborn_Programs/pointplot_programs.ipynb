```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q1
**Python Program to draw a point plot for sex against survived for a dataset given in url**

```python
def main():
    df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')
    
    # Create a point plot
    sns.pointplot(data=df, x='sex', y='survived', hue='sex', palette="deep", markers=["o", "s"], legend=True)
    
    # Set plot labels and title
    plt.title('Survival by Sex')
    plt.xlabel('Sex')
    plt.ylabel('Survived')
    plt.grid(axis='y', linestyle='--', alpha=0.9)
    plt.show()
    

if __name__ == "__main__":
    main()
```
