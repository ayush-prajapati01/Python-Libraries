```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q1
**Python Program to draw a scatter plot of “day” against “total bill” for a dataset given in a url**
<br><br>
Dataset: `https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv`

```python
def main():
    # Loading the dataset using the link
    df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
    print(f"The dataset is:\n{df.head()}\n")
    print(f"\nThe scatter plot is\n\n")
    plt.figure(figsize=(12,6))
    
    # Create a scatter plot
    sns.scatterplot(data=df, x='day', y='total_bill', hue='day', palette="deep", legend=True)
    
    # Set plot labels and title
    plt.title('Day vs Total Bill')
    plt.xlabel('Day')
    plt.ylabel('Total Bill')
    plt.grid(axis='y', linestyle='--', alpha=0.9)
    plt.legend(loc = "upper right")
    plt.show()
    

if __name__ == "__main__":
    main()
```
