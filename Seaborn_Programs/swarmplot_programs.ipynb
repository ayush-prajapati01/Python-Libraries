```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q1 
**Write a program to draw a swarm plot of total bill against size  for a  given dataset** 
<br><br>
Dataset: `https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv`


```python
def main():
    # Loading the dataset using the link
    df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
    print(f"The columns of the dataset are: \n{df}\n\n")
    print(f"The columns of the dataset are: {df.columns}")
    print(f"\nThe box plot is\n\n")
    plt.figure(figsize=(13,7))
    
    # Create a swarm plot
    sns.swarmplot(data=df, x='size', y='total_bill', hue='size', palette="Spectral", legend='auto')
    
    # Set plot labels and title
    plt.title('Size vs Total Bill')
    plt.xlabel('Size', labelpad=10)
    plt.ylabel('Total Bill', labelpad=12)
    plt.grid(axis='y', linestyle='-.', alpha=0.7)
    plt.legend(loc = "lower right")
    plt.show()
    

if __name__ == "__main__":
    main()
```

## Q2 
**Write a  program to draw swarm plot of “total bill” against day for a dataset given in url** 
<br><br>
Dataset: `https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv`


```python
def main():
    # Loading the dataset using the link
    df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
    print(f"The columns of the dataset are: \n{df}\n\n")
    print(f"The columns of the dataset are: {df.columns}")
    print(f"\nThe box plot is\n\n")
    plt.figure(figsize=(13,7))
    
    # Create a swarm plot
    sns.swarmplot(data=df, x='day', y='total_bill', hue='size', palette="flare", legend='auto')
    
    # Set plot labels and title
    plt.title('Day vs Total Bill')
    plt.xlabel('Day', labelpad=10)
    plt.ylabel('Total Bill', labelpad=12)
    plt.grid(axis='y', linestyle='-.', alpha=0.7)
    plt.legend(loc = "lower right")
    plt.show()
    

if __name__ == "__main__":
    main()
```
