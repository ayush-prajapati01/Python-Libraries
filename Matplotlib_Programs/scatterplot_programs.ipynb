```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1
**Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other.**

```python
def main():
    x_values = np.random.randint(100, size=(15))
    y_values = np.random.randint(200, size=(15))

    plt.scatter(x_values, y_values)
    plt.grid(True)
    plt.show()


if __name__ == "__main__":
    main()
```

## Q2
**Python program to draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other.**

```python
def main():
    x_values = np.random.randint(100, size=(25))
    y_values = np.random.randint(150, size=(25))

    plt.scatter(x_values, y_values, facecolors='none' , color = 'maroon')
    plt.grid(True)
    plt.show()


if __name__ == "__main__":
    main()
```

## Q3
**Python program to draw a scatter plot using random distributions to generate balls of different sizes.**

```python
def main():
    x_values = np.random.randint(100, size=(25))
    y_values = np.random.randint(150, size=(25))
    sizes = np.random.randint(20, 100, size=(25))
    colors = np.random.randint(100, size=(25))

    plt.scatter(x_values, y_values, marker = 'o' ,c = colors, cmap='BuPu_r', s = sizes)

    plt.title('Scatter Plot with different sizes')
    plt.grid(False)
    plt.show()


if __name__ == "__main__":
    main()
```
