```python
import numpy as np
import matplotlib.pyplot as plt
```

## Q1
**Python programming to create a pie chart of the popularity of programming Languages.**
<br><br>
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>

```python
def main():
    programming_languages = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])

    plt.pie(popularity)
    plt.show()


if __name__ == "__main__":
    main()
```

## Q2
**Python program to create a pie chart with a title of the popularity of programming Languages.**

```python
def main():
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    plt.pie(popularity, labels=programming_languages, startangle=90)
    plt.title("Popularity of Programming Languages")

    plt.show()


if __name__ == "__main__":
    main()
```
