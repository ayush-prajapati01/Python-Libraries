```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Q1
**Python programming to create a pie chart of the popularity of programming Languages.**
<br><br>
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>

```python
def main():
    programming_languages = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])

    plt.pie(popularity)
    plt.show()


if __name__ == "__main__":
    main()
```

## Q2
**Python program to create a pie chart with a title of the popularity of programming Languages.**

```python
def main():
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    plt.pie(popularity, labels=programming_languages, startangle=90)
    plt.title("Popularity of Programming Languages")

    plt.show()


if __name__ == "__main__":
    main()
```

## Q3
**Python programming to create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics.**

```python
def main():
    medals_df = pd.read_csv('medals.csv')
    print(f"The given data is:\n{medals_df}\n")
    explodes = [0.1,0,0,0,0]
    colors = ['#6600cc', '#0099cc', '#00e64d', '#cc5200', '#8000ff']

    plt.figure(figsize=(10,9))

    wedges, texts, autotexts = plt.pie(medals_df['gold_medal'], labels=medals_df['country'], autopct='%1.1f%%', startangle=90, 
            colors=colors, explode=explodes, shadow = False)

    # Method 1
    # for text, color in zip(texts, colors):
    #     text.set_color(color)

    # Method 2
    for wedge, text in zip(wedges, texts):
        text.set_color(wedge.get_facecolor())

    plt.title("2016 Olympic Top Five Most Successful Countries ")
    plt.legend(title = 'Country')
    plt.show()

if __name__ == "__main__":
    main()
```
