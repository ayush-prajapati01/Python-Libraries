```python
import numpy as np
import matplotlib.pyplot as plt
```

## Q1
**Python programming to display a bar chart of the popularity of programming Languages.**

```python
def main():
    programming_languages = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])

    plt.bar(programming_languages, popularity)
    plt.show()


if __name__ == "__main__":
    main()
```

## Q5
**Python program to display a bar chart of the popularity of programming Languages. Attach a text label above each bar displaying its popularity (float value).**

```python
def main():
    programming_languages = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    colors = ['#440066', '#d580ff']

    plt.figure(figsize=(10, 7))
    bars = plt.bar(programming_languages, popularity, color = colors)
    
    # Method 1
    # for i in range(len(popularity)):
    #     plt.text(i, popularity[i] + 0.5, popularity[i], ha='center')

    # Method 2
    # plt.bar_label(bars, fmt='%.1f', padding = 3)

    # Method 3
    plt.bar_label(bars, labels= [value for value in popularity], padding = 3)

    plt.xlabel('Progamming Languages')
    plt.ylabel('Popularity')
    plt.title('Programming Language Popularity')
    plt.show()


if __name__ == "__main__":
    main()
```

## Q7
**Python program to display a bar chart of the popularity of programming Languages (Specify the position of each bar plot).**

```python
def main():
    programming_languages = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    x_positions = [2,3,1,6,5,4]
    colors = ['#440066', '#d580ff']

    plt.figure(figsize=(8, 6))
    plt.bar(x_positions, popularity, color=colors, tick_label=programming_languages)

    plt.xlabel('Progamming Languages')
    plt.ylabel('Popularity')
    plt.title('Programming Language Popularity')
    plt.show()


if __name__ == "__main__":
    main()
```

## Q8
**Python program to display a bar chart of the popularity of programming Languages (Select the width of each bar and their positions).**

```python
def main():
    programming_languages = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    x_positions = np.array([2,1,4,3,6,5])
    widths = np.array([0.15, 0.25, 0.35, 0.45, 0.55, 0.65])
    colors = ['#440066', '#d580ff']

    plt.figure(figsize=(8, 6))
    plt.bar(x_positions, popularity, color=colors, width = widths, tick_label=programming_languages)

    plt.xlabel('Progamming Languages')
    plt.ylabel('Popularity')
    plt.title('Programming Language Popularity')
    plt.show()


if __name__ == "__main__":
    main()
```
