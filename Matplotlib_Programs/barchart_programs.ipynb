```python
import numpy as np
import matplotlib.pyplot as plt
```

## Q1
**Python programming to display a bar chart of the popularity of programming Languages.**

```python
def main():
    programming_languages = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])

    plt.bar(programming_languages, popularity)
    plt.show()


if __name__ == "__main__":
    main()
```

## Q2
**Python program to display a horizontal bar chart of the popularity of programming Languages.**

```python
def main():
    programming_languages = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])

    plt.barh(programming_languages, popularity)
    plt.show()


if __name__ == "__main__":
    main()
```
