## Q1
**Python programming to display a bar chart of the popularity of programming Languages.**

```python
def main():
    programming_languages = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])

    plt.bar(programming_languages, popularity)
    plt.show()


if __name__ == "__main__":
    main()
```

## Q9
**Python program to display a bar chart of the popularity of programming Languages (Increase bottom margin).**

```python
def main():
    programming_languages = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    colors = ['#440066', '#d580ff']

    plt.figure(figsize=(8, 6))
    plt.bar(programming_languages, popularity, color=colors)
    plt.subplots_adjust(bottom=0.4)

    plt.xlabel('Progamming Languages', labelpad = 10)
    plt.ylabel('Popularity', labelpad = 6)
    plt.title('Programming Language Popularity')
    plt.show()


if __name__ == "__main__":
    main()
```

## Q10
**Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.**
<br><br>
Sample Data:<br>
Means (men) = (22, 30, 35, 35, 26)<br>
Means (women) = (25, 32, 30, 35, 29)<br>

```python
def main():
    men_score = (22, 30, 35, 35, 26)
    women_score = (25, 32, 30, 35, 29)

    n = len(men_score)
    x_values = np.arange(n)
    width = 0.25

    plt.figure(figsize=(8,6))

    men_bars = plt.bar(x_values, men_score, width = 0.25, color = '#1a8cff', label = 'men')
    plt.bar_label(men_bars, labels = men_score, padding = 2)

    women_bars = plt.bar(x_values+width, women_score, width = 0.25, color = '#ff80ff', label = 'women')
    plt.bar_label(women_bars, labels = women_score, padding = 2)

    plt.title('Mens vs Women in each group')
    plt.xlabel('Groups', labelpad= 8)
    plt.ylabel('Scores', labelpad= 5)
    plt.xticks(x_values+width/2, ['Group A', 'Group B', 'Group C', 'Group D', 'Group E'])
    plt.legend()
    plt.show()


if __name__ == "__main__":
    main()
```
